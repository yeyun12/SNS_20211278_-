```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
bestbook_list = []

# 1~42페이지 반복
for page in range(1, 43):
    url = f"https://www.yes24.com/product/category/bestseller?categoryNumber=001&pageNumber={page}"
    res = requests.get(url, headers=headers)
    html_str = res.text

    print(f" {page}페이지 상태 코드: {res.status_code}")

    soup = BeautifulSoup(html_str, "html.parser")
    book_list = soup.select("li[data-goods-no]")

    for item in book_list:
        try:
            # 순위
            rank_tag = item.select_one("em.ico.rank")
            rank = rank_tag.text.strip() if rank_tag else "?"

            # 책 제목
            title_tag = item.select_one("a.gd_name")
            title = title_tag.text.strip() if title_tag else "?"

            # 저자
            author_tag = item.select_one("span.authPub.info_auth a")
            author = author_tag.text.strip() if author_tag else "?"

            # 출판사
            pub_tag = item.select_one("span.authPub.info_pub a")
            publisher = pub_tag.text.strip() if pub_tag else "?"

            # 출력
            print(f"[{rank}] {title} | 저자: {author} | 출판사: {publisher}")

            # 리스트 저장
            bestbook_list.append([rank, title, author, publisher])

        except Exception as e:
            print("에러:", e)
            continue

    print(f" {page}페이지 완료\n")
```

```python
# 수집 결과 확인
print(f" 총 수집된 책 수: {len(bestbook_list)}권")

# DataFrame 변환
df = pd.DataFrame(bestbook_list, columns=["순위", "책제목", "저자", "출판사"])

# CSV 저장
df.to_csv("sns기말과제.csv", index=False)
print("저장 완료")
```
